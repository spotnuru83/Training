#### Step 2 : Cleaning of Data

```python
df["TotalCharges"] = df["TotalCharges"].astype("float")
```

```python
df["TotalCharges"] = df["TotalCharges"].replace(' ',0)
```

```python
df["TotalCharges"] = df["TotalCharges"].astype("float")
```

```python
df.info()
```

```python
def conv(value):
    if value == 0:
        return "No"
    else:
        return "Yes"
```

```python
df["SeniorCitizen"] = df["SeniorCitizen"].apply(conv)
```

```python
df["SeniorCitizen"].value_counts()
```

#### Step 3: Plotting and Analyzing Data

```python
plt.figure(figsize=(3,5))
ax = sns.countplot(x="Churn",data=df)
ax.bar_label(ax.containers[0])
plt.show()
```

```python
gb = df.groupby("Churn").agg({"Churn":"count"})
```

```python
gb
```

```python
plt.figure(figsize=(3,3))
plt.pie(gb['Churn'],labels= gb.index, autopct = "%1.2f%%")
plt.title("Percentage of Churned Customers")
plt.show()
```

```python
plt.figure(figsize=(3,4))
ax = sns.countplot(x="gender",data =df, hue="Churn")
ax.bar_label(ax.containers[0])
plt.title("Gender vs Churn")
plt.show()
```

```python
def param_vs_churn(param):
    plt.figure(figsize=(3,4))
    ax = sns.countplot(x=param,data =df, hue="Churn")
    ax.bar_label(ax.containers[0])
    ax.bar_label(ax.containers[1])
    plt.title(f"{param} vs Churn")
    plt.show()
```

```python
param_vs_churn("Contract")
```

```python
param_vs_churn("SeniorCitizen")
```

```python
total_counts = df.groupby("SeniorCitizen")['Churn'].value_counts(normalize=True).unstack()*100
```

```python
fig,ax = plt.subplots(figsize=(4,4))

total_counts.plot(kind="bar", stacked=True, ax=ax, color=["#1f77b4","#ff7f0e"])
for p in ax.patches:
    width, height = p.get_width(),p.get_height()
    x,y = p.get_xy()
    ax.text(x+width/2, y+height/2,f'{height:.1f}%',ha='center',va='center')

plt.title("Stacked SeniorCitizen Churn Percentage")
plt.xlabel('SeniorCitizen')
plt.ylabel('Percentage (%)')
plt.xticks(rotation=0)
plt.legend(title='Churn', bbox_to_anchor=(0.9,0.9))

plt.show()
```

```python
columns  = ["PhoneService","MultipleLines","InternetService", "OnlineSecurity", "OnlineBackup", "DeviceProtection","TechSupport","StreamingTV","StreamingMovies"]

n_cols= 3
n_rows = (len(columns)+n_cols-1)//n_cols

fig,axes = plt.subplots(n_rows, n_cols, figsize=(15,n_rows*4))
axes = axes.flatten()

for i, col in enumerate(columns):
    ax = sns.countplot(x=col, data=df, ax=axes[i], hue=df["Churn"])
    ax.bar_label(ax.containers[0])
    ax.bar_label(ax.containers[1])
    axes[i].set_title(f"Count plot of {col}")
    axes[i].set_xlabel(col)
    axes[i].set_ylabel("Count")



plt.tight_layout()
plt.show()
    
```

```python
df_non_cat = df[["Churn_Int","TotalCharges","tenure","MonthlyCharges"]]
```

```python
df_corr = df_non_cat.corr()
```

```python
tgt_corr = df_corr['Churn_Int'].abs().sort_values(ascending=False)
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df_corr, annot=True, cmap='coolwarm', fmt='.2f',linewidth=0.5)
plt.title("Correlation Mapping")
plt.show()
```
